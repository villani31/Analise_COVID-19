# Analise do COVID-19 (Coronavírus)

A **COVID-19** é uma doença infecciosa causada pelo novo **coronavírus** (SARS-CoV-2) e tem como principal sintomas, febre, cansaço e tosse seca. 

É um vírus que caus uma doença respiratória, a COVID-19 pelo agente coronavírus, identificado em dezembro de 2019 na Chiva. Os coronavírus são uma grande familia viral, conhecida desde meados de 1960, que causam infecções respiratórias em seres humanos.

Os primeiros casos da doença **COVID-19** surgiram em Wuhan, uma cidade de 11 milhões de pessoas na província chinesa de Hubei, no final de 2019. Mais de 200 países já relataram casos do Covid-19, a doença provocada pelo coronavírus. A Organização Mundial de Saúde (OMS) declarou o surto como uma pandemia, que é uma epidemia que ganha escala global.

## Analise dos Dados

Os dados para essa analise foi obtido no repositório do Github da 
Our World in Data [owid](https://github.com/owid/covid-19-data/), para essa analise foi usado o arquivo owid-covid-data.csv.

O Dataset passa por atualização todos os dias, as visualização nesse projeto, estão com base no dia 31-08-21, caso seja reproduzido esse projeto, em outro momento, os valores poderão ser diferentes.

Importante: As visualizações das analise, é feita conforme são inserido os dados no dataset, pode ocorrer casos que os dados foram inserido com menas frequências em uma determinada data, com isso a visão da evolução não fica clara, como deveria ser.

Vamos ao projeto:

```python
# importar as bibliotecas necessarias
import pandas as pd
import matplotlib.pyplot as plt

# importar dataset do COVID-19 para um Dataframe
DATA_PATH = 'https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv'
df = pd.read_csv(DATA_PATH, index_col='date', parse_dates=True)

# dimensões do Dataframe
print('Dimensões do Dataframe: \n',
      'Numero de variaveis: ', df.shape[1], '\n',
      'Numero de linhas: ', df.shape[0])

# lendo as 5 primeiras linhas do Dataframe
df.head()
```

```python
# lendo as 5 ultimas linhas do Dataframe
df.tail()
```

* O comando a baixo é um método que imprime informações sobre um dataframe, quantidade de linhas com valores e valores ausentes, e o tipo das variaveis.

```python
# extrair informacoes do Dataframe
df.info()
```

* Calculo em % dos 20 colunas em ordem descrescente, com maior número de valores ausente em cada variável, em determinado casos, dependendo da quantidade, pode ser um problema, precisa ser tratado antes de continuar com a analise, no nosso caso, como não vai ser feito uma analise profunda nos dados, iremos somente analisar casos e mortes confirmados, não irá interfeirir.

```python
# identificr a porcentagem de valores ausente das colunas, listar 20 colunas por orgem decrescente
((df.isnull().sum() / df.shape[0]).sort_values(ascending=False) * 100)[:20]
```

* Como o dataset tem muitas coluna, foi criado um novo dataframe, que contem somente as colunas que será usando na analise referente ao COVID-19.

```python
# novo dataframe selecionando as colunas
df_analise = df[['iso_code','continent','location','new_cases','new_deaths','total_cases','total_deaths',
    'people_fully_vaccinated','gdp_per_capita']]
```

```python
# ver as primeiras 5 linhas
df_analise.tail()
```

## Quais paises tem mais casos de mortes até a data atual

Foi plotado um gráfico de histograma, que analisou os 5 paisies com maior número de mortes, analisando os dados inseridos até a data atual 31-08-2021, o Brasil, tem a segunda posição de maior número de morte no mundo.

Importante: Essa analise é embasado nos dados inseridos no Dataset, pode ser que esse valor não é seja real, pois tem países, como por exemplo a China, nãoregistou o número correto de mortes.

```python
# Quais paises tem mais casos de mortes até a data atual
df_morta_data_atual = (df_analise.loc[df_analise.index == '2021-08-31'].sort_values(by='total_deaths', 
                                                                                    ascending=False)).head(11)

# listando Dataframe filtrado
df_morta_data_atual[['continent','location','total_deaths']].iloc[6:]
```

* Podem ver, o Brasil com mais de 580 mil mortes confirmada.

```python
# plotar gráfico de barras com 5 paises com mais mortes
df_morta_data_atual_plot = df_morta_data_atual[['continent','location','total_deaths']].iloc[6:]
fig, ax = plt.subplots()
ax.bar('location', 'total_deaths' ,data=df_morta_data_atual_plot)
ax.set_title('Os 5 paises com mais morte por COVID-19')
plt.show()
```

## Total de casos no mundo

Plot de um gráfico de linha, de casos confirmados e mortes que ocorreu pelo mundo, no primeiro gráfico, de casos confirmados pelo mundo, se percebe que entre os meses 06 e 07, teve uma queda nos casos confirmados.

Casos confirmados no mundo mais de 219 mi, e morte mais de 4 mi.

```python
# plot com total de casos confirmados
fig, ax = plt.subplots(figsize=(15,4))
df_analise['total_cases'].plot(kind='line', x=ax)
ax.set_title('Evolução de casos confirmados no mundo')
ax.set_xlabel('Data')
plt.show()
```

```python
# plot evolucao de mortes pelo mundo
fig, ax = plt.subplots(figsize=(15,3))
df_analise.loc[df_analise.location == 'World'].plot(kind='line', y='total_deaths', ax=ax)
#df_analise.loc[df_analise.location == 'World','total_deaths'].reset_index(drop=True).plot(kind='line', ax=ax)
ax.set_title('Evolução de mortes pelo mundo')
plt.show()
```

## Total de casos no Brasil

Seguindo a mesma linha da analise feita nos casos pelo mundo, foi feito uma analise somente para o Brasil, de casos e mortes confirmados.

Casos confirmados mais de 20 mi, e morte confirmadas mais de 580 mil.

```python
# plot total de casos confirmados e mortes no Brasil
fig, (ax1,ax2) = plt.subplots(nrows=2, ncols=1, figsize=(10,5))
df_analise.loc[df_analise.location == 'Brazil'].plot(kind='line', y=['total_cases'], ax=ax1, )
df_analise.loc[df_analise.location == 'Brazil'].plot(kind='line', y=['total_deaths'], color='r', ax=ax2)
ax1.set_title('Total de casos confirmado no Brasil')
ax2.set_title('Total de mortes no Brasil')
ax1.set_xlabel('Data')
ax2.set_xlabel('Data')
plt.tight_layout()
```

## PIB dos paises pelo número de mortes

O PIB (produto interno bruto) é a soma de todos os bens e serviços finais produzidos por um país, estado ou cidade, geralmente em um ano, a baixo um plot de total de mortes X renda per capita.

```python
# grafico considerando PIB dos paises pelo numero de mortos
fig, ax = plt.subplots(figsize=(8,5))
df_analise.loc[df_analise.index == '2021-08-31'].plot.scatter(x='total_deaths', 
                                                              y='gdp_per_capita', ax=ax , grid=True)
#df_analise.plot.scatter(x='total_deaths', y='gdp_per_capita', ax=ax , grid=True)
ax.set_title('Total de mortes X Renda Per Capita em 31-08-2021')
plt.tight_layout()
```

## Gráfico de casos da cidade de Curitibanos - SC

Foi feito uma analise rápida do dataset do estado de santa catarina, para cidade de Curitibanos - SC, o dataset foi baixado do site [Brasil.io](https://brasil.io/dataset/covid19/files/), o Brasil.IO tem como objetivo facilitar o acesso a dados públicos brasileiros, esses dados são inserido através de colaboração das pessoas.

Curitibanos é um município brasileiro do estado de Santa Catarina, que possui população de 39.893 mil habitantes.

```python
# importar dataset para Dataframe do estado de santa catarina
DATA_PATH_SC = 'caso.csv'
df_SC = pd.read_csv(DATA_PATH_SC, index_col='date', parse_dates=True)
```

```python
# filtrando dados para cidade de Curitibanos
df_curitibanos = df_SC[['state','city','confirmed','deaths']].loc[df_SC.city == 'Curitibanos'].head()
df_curitibanos.head()
```

```python
# lendo as ultimas linhas do dataframe filtrado
df_curitibanos.tail()
```

* Com mais de 5 mil casos confirmados, e 181 mortes.

```python
# plot de casos para cidade de Curitibanos - SC
fig, (ax1,ax2) = plt.subplots(nrows=2, figsize=(10,5))
df_curitibanos.confirmed.plot(kind='line', ax=ax1)
df_curitibanos.deaths.plot(kind='line', color='r', ax=ax2)
ax1.set_title('Casos confirmados para cidade de Curitibanos - SC')
ax2.set_title('Mortes confirmado para cidade de Curitibanos - SC')
ax1.set_xlabel('Data')
ax2.set_xlabel('Data')
plt.tight_layout()
```
